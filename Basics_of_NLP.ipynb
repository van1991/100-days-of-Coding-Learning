#### Tokenize Funtion


```
# Define a function to tokenize any text

def tokenz(text):
   import spacy
   spacy_nlp = spacy.load('en_core_web_sm')
   doc = spacy_nlp(text)
   tok = [token.text for token in doc]
   from nltk.probability import FreqDist
   fdist = FreqDist(tok)

   ## using nltk
   import nltk
   nltk.download("popular")
   words = nltk.word_tokenize(req)
   gdist = FreqDist(words)
   gdist.plot(15)
   fdist.plot(15) # Comparing Spacy output with NLTK
   return

tokenz(req)
```

#### How to remove Punctuations from text?
##### In Python, isalpha() is a built-in method used for string handling. The isalpha() methods returns True if all characters in the string are alphabets, Otherwise, It returns False. This function is used to check if the argument includes only alphabet characters 

```
# Steps to remove punctuations as they are the post frequent ones in the text - here frequency for [.] is 9.

# Empty list to store words
words_no_punc = []

# For remvoing punctuation marks :
for w in words:
  if w.isalpha() :
    words_no_punc.append(w.lower())

print("New length of the text after removing punctuations",len(words_no_punc))
print("\n")
print(words_no_punc)
```

```
# define a function to show word frequency and plot

def words_freq_plot (word_list):
    # import required libraries :
    from nltk.probability import FreqDist
    fdist = FreqDist(word_list)
    print(fdist.most_common(15))
    fdist.plot(15)
# Now call the function and input words_no_punc
words_freq_plot(words_no_punc)
```

## How to remove Stop Words?

```
from nltk.corpus import stopwords

# List of stopwords
stopwords = stopwords.words("english")
print("There are ",len(stopwords)," stopwords in english")
print(stopwords)
```

```
# Empty list to store clean words
clean_words = []

for w in words_no_punc :
   if w not in stopwords :
     clean_words.append(w)

print("New length of the text after removing stopwords",len(clean_words))
print("\n")
print(clean_words)
```

```
# Now call the function and input clean_words to see Frequency Distribution
words_freq_plot(clean_words)
```

## Stemming of the Text

```
# Import the stemming library
from nltk.stem import PorterStemmer

porter =  PorterStemmer()

# Create a word list 
word_list =  ["Study","Studying","Studies","Studied"]

for w in word_list:
  print(porter.stem(w))
```

There are multiple way to do stemming but it is not a good way to clean the data. Better approach is Lemmatization. But challenge with Lemmatization is that it is time consuming.

```
import nltk
nltk.download('wordnet')

from nltk import WordNetLemmatizer
lemma = WordNetLemmatizer()
word_list =  ["Study","Studying","Studies","Studied","am","is","are","was","were","studies","leaves","decreases","plays"]
print(word_list)

for w in word_list:
    print(lemma.lemmatize(w ,pos="v"))
  
```

```
#The default value of PoS in lemmatization is a noun(n).
print(lemma.lemmatize('studying',pos = 'v'))
print(lemma.lemmatize('studying',pos = 'n'))
print(lemma.lemmatize('studying',pos = 'a'))
print(lemma.lemmatize('studying',pos = 'r'))
```

## Named Entity Recognition (NER)
###### Named entity recognition can automatically scan entire articles and pull out some fundamental entities like people, organizations, places, date, time, money, and GPE discussed in them.


```
# An example of a sentence
sentence = "Mr. Smith made a deal on a beach of Switzerland near WHO"

# first tokenize the sentence
tokenized_words = word_tokenize(sentence)
print(tokenized_words)

import pandas as pd
h =  pd.DataFrame(tokenized_words)

# then POS tag them
for w in tokenized_words:
  tagged_words = nltk.pos_tag(tokenized_words)

g = pd.DataFrame(tagged_words)
#print(g)


# Named Entity Regoznition
N_E_R = nltk.ne_chunk(tagged_words,binary=False)
i = pd.DataFrame(N_E_R)
#print(i)

print(pd.concat([g,i],axis=1))
```

## WordNet
#### Wordnet is a lexical database for the English language. Wordnet is a part of the NLTK corpus. We can use Wordnet to find meanings of words, synonyms, antonyms, and many other words.

```
# import wordnet
from nltk.corpus import wordnet

for words in wordnet.synsets("fun"):
  print (words)
```

#### Word meanings with definitions

```
# Word meanings with definitions

for words in wordnet.synsets("singing"):
  print(words.name())
  print(words.definition())
  print(words.examples())

  for lemma in words.lemmas():
    print(lemma)
  print("\n")
```

#### Hypernyms: Hypernyms gives us a more abstract term for a word.
 Hyponyms: Hyponyms gives us a more specific term for a word.

```
print((wordnet.synsets("Singing")[0]).hypernyms())
print((wordnet.synsets("Singing")[0]).hyponyms())
print((wordnet.synsets("Singing")[0]).lemmas())
print((wordnet.synsets("Singing")[0]).lemmas()[1].name()) # Get only name
```

## Bag of Words
###### It is a method of extracting essential features from row text so that we can use it for machine learning models. We call it Bag of words because we discard the order of occurrences of words. A bag of words model converts the raw text into words, and it also counts the frequency for the words in the text. In summary, a bag of words is a collection of words that represent a sentence along with the word count where the order of occurrences is not relevant.

### Bag of Words implementation

```
# import required libraries
from sklearn.feature_extraction.text import CountVectorizer

# Text for analysis
sentences = ["Jim and Pam travelled by the bus:",
             "The train was late",
             "The flight was full.Travelling by flight is expensive"]
#print(sentences)

# Create an object
cv = CountVectorizer()

# Generating output for Bag of Words
B_O_W = cv.fit_transform(sentences).toarray()
print(B_O_W)

# Total words in the BOW with their Index in the model
print(cv.vocabulary_)
print("\n")

# Extract only the words w/o index

col = cv.get_feature_names()

import pandas
df = pandas.DataFrame(B_O_W)
df.columns = col # assign column names through a list
df['Sentence'] = sentences
print(sentences)
df
```
